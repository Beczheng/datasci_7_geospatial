# 4. Cleaning the datasets

## Dataset #1

```python
# Cleaning columns in dataset #1

def clean_column(df):
  def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

  df.columns = [clean_name(col) for col in df.columns]
  return df

new_df1 = clean_column(df1)
new_df1
```

# 5. Graphs and insights

## Dataset #1

```python
# Graph for dataset #1

fig, ax = plt.subplots(1, 1)
divider = make_axes_locatable(ax)

new_df1.plot(
    ax=ax,
    cax=divider.append_axes("bottom", size="5%", pad=0.1),
    cmap="Reds",
    column="teenbirth",
    legend_kwds={"label": "Teen Birth Rates", "orientation": "horizontal"},
    legend=True,
);
```

**Description of Dataset:** This dataset contains information on births rates across Lake County, Illinois. The graph above specifically shows teen birth rates by zip code. Teen births are defined as women aged 15 to 19 years who give birth.

**Patterns or Anomalies:** Teen birth rates are medium to high around the the upper half of Lake County.
